```python
from sklearn.datasets import load_svmlight_file
import numpy as np
import random
import math

# 读取数据
X_train, y_train = load_svmlight_file('a9a.txt', n_features = 123)
X_test, y_test = load_svmlight_file('a9a_testing.txt', n_features = 123)

X_train = X_train.toarray()
X_test = X_test.toarray()
y_train = y_train.reshape(y_train.shape[0], 1)
y_test = y_test.reshape(y_test.shape[0], 1)
```

```python
INDEX_LIST = range(0, X_train.shape[0])
# 定义采样函数
def select_samples(X, y, n):
    samples_index = random.sample(INDEX_LIST, n)
    X_samples = np.ones((0, X.shape[1]))
    y_samples = np.ones((0, y.shape[1]))
    
    for item in samples_index:
        X_samples = np.r_[X_samples, X[item].reshape(1, X.shape[1])]
        y_samples = np.r_[y_samples, y[item].reshape(1, y.shape[1])]
    
    return X_samples, y_samples
```

```python
# 计算梯度的函数
def compute_gradient(X, y, W):
    loss_gradient = np.zeros(W.shape)
    for index in range(0, y.shape[0]):
        Xi = X[index].reshape(1, X.shape[1]).T
        Yi = y[index][0]
        
        loss_gradient += Yi / (1 + math.exp(Yi * np.dot(W.T, Xi)[0][0])) * Xi    
        
    gradient = LAMBDA * W - 1 / batch_num * loss_gradient
    
    return gradient
```

```python
# NAG
def NAG(X, y, W, params):
    W_ = W - params['GAMMA'] * params['Vt']
    
    gradient = compute_gradient(X, y, W_)
    params['Vt'] = params['GAMMA'] * params['Vt'] + params['ETA'] * gradient
    W = W - params['Vt']
    
    return W, params

# RMSProp
def RMSProp(X, y, W, params):
    EPSILON = 1e-8
    
    gradient = compute_gradient(X, y, W)
    params['Gt'] = params['GAMMA'] * params['Gt'] + (1 - params['GAMMA']) * (gradient * gradient)
    W = W - (params['ETA'] / np.sqrt(params['Gt'] + EPSILON)) * gradient
    
    return W, params

# AdaDelta
def AdaDelta(X, y, W, params):
    EPSILON = 1e-8
    GAMMA = params['GAMMA']
    
    gradient = compute_gradient(X, y, W)  
    params['Gt'] = GAMMA * params['Gt'] + (1 - GAMMA) * (gradient * gradient)
    Delta_W = -(np.sqrt(params['Delta_t'] + EPSILON) / np.sqrt(params['Gt'] + EPSILON)) * gradient
    params['Delta_t'] = GAMMA * params['Delta_t'] + (1 - GAMMA) * (Delta_W * Delta_W)
    W = W + Delta_W
    
    return W, params

# Adam
def Adam(X, y, W, params):
    EPSILON = 1e-8
    BETA1 = params['BETA1']
    GAMMA = params['GAMMA']
    
    gradient = compute_gradient(X, y, W)
    params['mt'] = BETA1 * params['mt'] + (1 - BETA1) * gradient
    params['Gt'] = GAMMA * params['Gt'] + (1 - GAMMA) * (gradient * gradient)
    ALPHA = params['ETA'] * math.sqrt(1 - GAMMA ** params['iteration']) / (1 - BETA1 ** params['iteration'])
    W = W - ALPHA * (params['mt'] / np.sqrt(params['Gt'] + EPSILON))
    
    return W, params
```

```python
# 初始化部分参数
LAMBDA = 1
batch_num = 100
threshold = 0.5
iteration_time = 100
```

```python
# 定义保存训练集损失值、验证集损失值和正确率的列表
NAG_loss_list = []
RMSProp_loss_list = []
AdaDelta_loss_list = []
Adam_loss_list = []
# 使用字典存储，方便后面把loss添加到相应的列表
loss_list = {'NAG': NAG_loss_list, 
             'RMSProp': RMSProp_loss_list, 
             'AdaDelta': AdaDelta_loss_list,
             'Adam': Adam_loss_list
             }

NAG_right_rate_list = []
RMSProp_right_rate_list = []
AdaDelta_right_rate_list = []
Adam_right_rate_list = []
right_rate_list = {'NAG': NAG_right_rate_list, 
             'RMSProp': RMSProp_right_rate_list, 
             'AdaDelta': AdaDelta_right_rate_list,
             'Adam': Adam_right_rate_list
             }

method_list = [NAG, RMSProp, AdaDelta, Adam]
method_names = ['NAG', 'RMSProp', 'AdaDelta', 'Adam']
```

```python
for method_index in range(0, int(len(method_list))):
    W = np.zeros((X_train.shape[1], 1))
    method_params = {'Vt': np.zeros(W.shape),
                 'Gt': np.zeros(W.shape),
                 'Delta_t': np.zeros(W.shape),
                 'mt': np.zeros(W.shape), 
                 'ETA': 0.01,
                 'GAMMA': 0.9,
                 'BETA1': 0.9,
                 'iteration': 1,
                 }
    if (method_list[method_index] == AdaDelta):
        method_params['GAMMA'] = 0.9999

    for item in range(0, iteration_time):
        X_samples, y_samples = select_samples(X_train, y_train, batch_num)
        W, method_params = method_list[method_index](X_samples, y_samples, W, method_params)
        method_params['iteration'] = item + 1
        
        # 计算验证集上的损失值和正确率并保存到列表中
        loss_sum = 0
        right_counter = 0
        for index in range(0, y_test.shape[0]):
            Xi = X_test[index].reshape(1, X_test.shape[1]).T
            Yi = y_test[index][0]
            
            WTXi = np.dot(W.T, Xi)[0][0]
            loss_sum += math.log(1 + math.exp(-Yi * WTXi))
            
            hw = 1 / (1 + math.exp(-WTXi))
            predict = -1
            if (hw > threshold):
                predict = 1
                
            if (predict == Yi):
                right_counter += 1
                
        right_rate = right_counter / y_test.shape[0]
        right_rate_list[method_names[method_index]].append(right_rate)
        
        test_loss = LAMBDA * 1/2 * np.dot(W.T, W)[0][0] + 1/X_test.shape[0] * loss_sum
        loss_list[method_names[method_index]].append(test_loss)
```

```python
# 打印正确率
for method_index in range(0, int(len(method_list))):
    name = method_names[method_index]
    print("The right rate list of " + name + ": ", right_rate_list[name])
```

```python
# 将从训练集和验证集上得到的损失值列表画成图像
import matplotlib.pyplot as plt

plt.xlabel('iteration time')
plt.ylabel('loss')

x = range(iteration_time)
plt.plot(x, NAG_loss_list, 'b-', label = 'NAG')
plt.plot(x, RMSProp_loss_list, 'k-', label = 'RMSProp')
plt.plot(x, AdaDelta_loss_list, 'r-', label = 'AdaDelta')
plt.plot(x, Adam_loss_list, 'y-', label = 'Adam')

plt.grid(True)
plt.legend()
plt.show()
```
